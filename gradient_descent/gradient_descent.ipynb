```python
import numpy as np
from matplotlib import pyplot as plt
import math
```

```python
m =100
b = 1.5
k = 5
#y = k * x + b
x = np.arange(m)/m
y = k*x + b + 0.1*np.random.randn(m)
```

```python
theta0 = 0.5
theta1 = 0.5
alpha = 0.1
ite = 1
eps = 10**(-10)
# temp0 = theta0 - alpha * 1/m * np.sum(theta0+theta1*x-y)
while True:
    temp0 = theta0 - alpha * 1/m * np.sum(theta0+theta1*x-y)
    temp1 = theta1 - alpha * 1/m * np.sum((theta0+theta1*x-y)*x)
    if np.abs(theta0 - temp0) < np.abs(eps and theta1 - temp1) < eps :
        theta0 = temp0
        theta1 = temp1
        break
    theta0 = temp0
    theta1 = temp1
    ite = ite + 1

print("iteration: ", ite)
print("theta0: ", theta0, "theta1: ", theta1)
y_estimate = theta0 + theta1 * x

plt.scatter(x,y)
plt.plot(x,y_estimate,linewidth='2',color='red')
plt.show()
```
